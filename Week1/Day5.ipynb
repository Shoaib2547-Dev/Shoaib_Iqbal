```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

df=pd.read_csv('/content/Titanic-Dataset.csv')
df
```

```python
sns.histplot(data=df,x='Age',kde=True,bins=20)
plt.show()
```



*   It count the how many people fall in certain age
bin
*   x-axis shows the Age and and make the 20 bins range from 0 to 80 of equal size 10 each


*   Data is continous no gap between bars












```python
sns.barplot(data=df,x='Sex',y='Age',hue='Survived')
plt.show()
```




*   It displays the average age of Titanic passengers, grouped by sex and whether they survived (0 for No, 1 for Yes).
*   Men: Those who survived were, on average, younger than those who died.

*  Women: Those who survived were, on average, older than those who died.

*   a categorical column for the x-axis and a numeric column for the y-axis, it automatically groups the data by category










```python
sns.boxplot(data=df,x='Sex',y='Age',hue='Survived')
plt.show()
```



*   The white Dots above the plots is outliers
*   The male age 80 is survived is outlier





```python
correlation=df.corr(numeric_only=True)

plt.title("Correlation Heatmap")
sns.heatmap(
    correlation,
    annot=True,
    fmt=".2f",
    cmap="coolwarm",
    linewidths=1,
    linecolor="white"
)

plt.show()

```



*   Passengers in 1st Pclass and those who paid higher Fares had a better chance of surviving.


*   Same colors show that they are highly correlated means one up ,other also up if they are both positive if negative then they are weakly correalted

*   White color shows that they are not linked
