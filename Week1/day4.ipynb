```python
import pandas as pd

Df_T=pd.read_csv("/content/sample_data/Titanic-Dataset.csv")
Df_T
```

Check Which Column have empty values

```python
Df_T.isnull().sum()
```

Fill the Age column missing values with mean

```python
Df_T['Age']=Df_T['Age'].fillna(Df_T['Age'].mean())
Df_T['Age']
```

For categorical data we use mode commonly

```python
Df_T['Cabin']=Df_T['Cabin'].fillna(Df_T['Cabin'].mode()[0])
Df_T['Cabin']
```

For categorical data we use mode commonly

```python
Df_T['Embarked']=Df_T['Embarked'].fillna(Df_T['Embarked'].mode()[0])
Df_T['Embarked']
```

Check Datatype

```python
Df_T.dtypes
```

Check Duplicate and remove if any

```python
Df_T.drop_duplicates()
```

Group By On basis of Survived or not With Age mean and Fare mean

```python
Df_T.groupby('Survived').agg({
    'Age':'mean',
    'Fare':'mean'
})
```

Merge to Label Survived (0,1) as Survived or not survived

```python
Survived={
    'Survived':[0,1],
    "Survived_Status":["Not Survived","Survived"]
}
Df_Survived=pd.DataFrame(Survived)
Df_Survived

Df_Merged=pd.merge(Df_T,Df_Survived,how='left')
Df_Merged
```
