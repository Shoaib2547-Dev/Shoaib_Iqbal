```python
import numpy as np
from numpy import random
from numpy.lib.stride_tricks import sliding_window_view

Nparray=np.round(20*np.random.rand(100))

def Rollingwindow():
    window=3
    windows=sliding_window_view(Nparray,window)

    for x in windows:
        W_mean=x.mean()
        print(x)
        print(W_mean)
        print(" ")
        print(" ")


Rollingwindow()
```

```python
def normalizeData():

    std=np.std(Nparray)
    mean=np.mean(Nparray)

    print("Z_score")
    z_score=Nparray-mean/std
    print("length",len(z_score))
    print()
    print(z_score)


    print()
    print("Outliers")
    Outliers=z_score[np.abs(z_score)>2]
    print("length",len(Outliers))
    print()
    print(Outliers)





normalizeData()


```
