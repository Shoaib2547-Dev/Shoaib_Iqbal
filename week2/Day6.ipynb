```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd


df=pd.read_csv('/content/Housing.csv')
df
```

```python
avg_price = df.groupby("bedrooms")["price"].mean()

plt.figure(figsize=(6,4))

plt.plot(avg_price.index, avg_price.values, marker="o")

plt.title("Average House Price by Bedrooms")
plt.xlabel("Bedrooms")
plt.ylabel("Average Price")

plt.grid(True)

plt.show()
```

**Line chart**
*   This chart shows how the average house price changes as the number of bedrooms increases.
*   Houses with more bedrooms generally have higher average prices.

*   this shows house with few bedrooms have less pirce,price incraese with bedroms increase






```python
avg_price = df.groupby("furnishingstatus")["price"].mean()

plt.figure(figsize=(6,4))

plt.bar(avg_price.index, avg_price.values)

plt.title("Average Price by Furnishing Status")
plt.xlabel("Furnishing Status")
plt.ylabel("Average Price")

plt.show()
```

***Bar Chart***
*   This chart compares the average price of furnished, semi-furnished, and unfurnished houses using groupby and take mean.
*   Furnished houses usually have higher average prices.
*   Less Furnished have generally less price than furnished house





```python
plt.figure(figsize=(6,4))

plt.hist(df["price"], bins=15)

plt.title("Distribution of House Prices")
plt.xlabel("Price")
plt.ylabel("Number of Houses")

plt.show()
```

**Histogram**
*   the chart shows how house prices are distributed.
*   Most houses fall within a certain price range usually less expensive

*   Few houses are very expensive






```python
plt.figure(figsize=(5,4))

plt.boxplot(df["price"])

plt.title("Boxplot of House Prices")
plt.ylabel("Price")

plt.show()
```

**Box:**
*   The boxplot shows the spread of house prices and the median price
*   Dots outside the box represent outliers(expensive).


*   Most of the datapoints are under 1




```python
plt.figure(figsize=(6,4))

plt.scatter(df["area"], df["price"], label="Area vs House Peice")

plt.legend()

plt.title("Area vs House Price")
plt.xlabel("Area")
plt.ylabel("Price")

plt.show()
```


*   **Scatter:**

*   This chart shows the relationship between house area and price.
*   Larger houses generally tend to have higher prices.
*   where datapoints are closer to each other shows that prices are almost same at these points







```python
corr = df.corr(numeric_only=True)

plt.figure(figsize=(8,6))

# in seaborn sns.heatmap()) is used but p=matplot imshow( ) for heatmap
plt.imshow(corr, cmap="coolwarm")
plt.colorbar()

plt.xticks(range(len(corr.columns)), corr.columns, rotation=45)
plt.yticks(range(len(corr.columns)), corr.columns)

plt.title("Correlation Heatmap")

plt.show()
```

**Heatmap**
*   The heatmap shows that they are related to each other or not by using correlation.
*   Values closer to 1 indicate a strong positive relationship, while values closer to -1 indicate a strong negative relationship.
*   Diagonal are 1 because same attribute are face to face
