```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd


df=pd.read_csv('/content/Titanic-Dataset.csv')
df
```

```python
df2 = df.sort_values("Age")

plt.plot(df2["Age"], df2["Fare"],marker='.')

plt.title("Age vs Fare")
plt.xlabel("Age")
plt.ylabel("Fare")

plt.show()
```

**Line Chart**
It shows how ticket fares change as passenger age increases.

```python
pclass = df["Pclass"].value_counts()


plt.bar(pclass.index, pclass.values)

plt.title("Passengers in Each Class")
plt.xlabel("Passenger Class")
plt.ylabel("Number of Passengers")

plt.show()
```

**Bar:**
It compares how many passengers traveled in each class.

```python
plt.hist(df["Age"].dropna())

plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")

plt.show()
```

It shows how passenger ages are distributed.

```python
plt.boxplot(df["Fare"])

plt.title("Fare")

plt.show()
```

**Box:**It helps identify the spread of ticket fares and any outliers.

```python
plt.scatter(df["Age"], df["Fare"],label="Age vs Fare")

# legend show labels
plt.legend()

plt.title("Age vs Fare")
plt.xlabel("Age")
plt.ylabel("Fare")

plt.show()
```


*   **Scatter:**It shows whether there is a relationship between age and ticket fare.
*   It also show that data has noise and outlier



```python
corr = df.corr(numeric_only=True)

# in seaborn sns.heatmap()) is used but p=matplot imshow( ) for heatmap
plt.imshow(corr)

plt.colorbar()

plt.xticks(range(len(corr.columns)), corr.columns, rotation=90)
plt.yticks(range(len(corr.columns)), corr.columns)

plt.title("Correlation Heatmap")

plt.show()
```

It shows how strongly the numerical columns are related to each other.
